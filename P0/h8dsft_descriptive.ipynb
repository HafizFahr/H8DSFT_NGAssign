# Descriptive statistics of COVID-19 Indonesia
find the measure of central tendency and measure of variance from [this](https://www.kaggle.com/hendratno/covid19-indonesia) dataset.

Import pandas and matplotlib.pyplot for data processing and visualization

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

Load data in a dataframe using read_csv() and display to check if it's loaded

```python
dt = pd.read_csv('c19_dataset.csv')
display(dt)
```

Let's focus on Jawa Barat. Create new dataframe and select only 'Location ISO Code' and 'New Cases' columns. Next, select ID-JB and drop other values.

```python
cln = dt[['Location ISO Code','New Cases']]
cln = cln[cln['Location ISO Code'] == 'ID-JB']
cln
```

Calculate the measures of central tendency (mean, mode, median)

```python
mn = cln['New Cases'].mean()
mo = cln['New Cases'].mode()
me = cln['New Cases'].median()
print(
    'Mean   :',mn,
    '\nMode   :',mo[0],
    '\nMedian :',me
)
```

Calculate the measure of variance (range, variance, standard deviation)

```python
ran = cln['New Cases'].max() - cln['New Cases'].min()
var = cln['New Cases'].var()
stdev = cln['New Cases'].std()
print(
    'Range    :',ran,
    '\nVariance :',var,
    '\nSt. Dev  :',stdev
)
```
